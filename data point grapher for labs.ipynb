```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy as sc
from scipy.optimize import curve_fit

#prep libraries

gen = [i for i in range(400, 751, 50)]
spec = [i for i in range(500,601, 20)]
rdx = sorted(list(set(gen + spec)))

#this sets our variable datapoints

uiy = input(f"Enter {len(rdx)} y values separated by spaces: ")
rdy = list(map(float, uiy.split()))

#user inpot datapoints
#mine are -0.1 0.1 0.12 0.17 0.19 0.19 0.18 0.16 0.13 0.06 0.02 0

# Check if the number of y values matches the number of x values
if len(rdy) != len(rdx):
    print(f"Error: The number of y values ({len(rdy)}) does not match the number of x values ({len(rdx)}). Please enter {len(rdx)} y values.")
else:
    # Define a quadratic function to fit
    def quadratic_function(x, a, b, c):
        return a * x**2 + b * x + c

    # Use curve_fit to find the parameters of the quadratic function that best fit the data
    popt, pcov = curve_fit(quadratic_function, rdx, rdy)

    # Generate y values based on the fitted quadratic function for plotting the trend line
    x_new = np.array(rdx)
    y_extrapolated = quadratic_function(x_new, *popt)


    plt.scatter(rdx, rdy, label='Observed Data')
    plt.plot(rdx, rdy, linestyle='-', color='blue', label='Connected Data Points') # Add this line to connect the data points
    plt.plot(x_new, y_extrapolated, color='red', label='Fitted Trend Line')
    plt.xlabel('wavelength') # Add x-axis label
    plt.ylabel('absorbance') # Add y-axis label
    plt.legend()
    plt.show()
```
